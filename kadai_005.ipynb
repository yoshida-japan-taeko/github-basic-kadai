```python
# 上辺・下辺・高さを変数に代入
top = 10      # 上辺（cm）
bottom = 20   # 下辺（cm）
height = 5    # 高さ（cm）

# 台形の面積を計算して変数 area に代入
# （上辺 + 下辺）× 高さ ÷ 2
area = (top + bottom) * height / 2

# 数値 + 文字列を結合して出力
print(str(area) + "cm²")
```
